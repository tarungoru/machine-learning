Multiple linear regression is a regression model that estimates the relationship between a quantitative dependent variable and two or more independent variables using a straight line

Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

importing the dataset

```python
data_set=pd.read_csv('Multiple Linear Regr.csv')
x=data_set.iloc[:,:-1]
y=data_set.iloc[:,-1]
print(x)
```

Encoding the categorial data
Categorical variables represent types of data which may be divided into groups. Examples of categorical variables are race, sex, age group, and educational level.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
x=np.array(ct.fit_transform(x))
print(x)
```

splitting the data into training set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)

```

Training the multiple linear regression


```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

predict the test results

```python
y_pred=reg.predict(x_test)
np.set_printoptions(precision=3)
y_test=np.array(y_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Visualizing the results

```python
pt.plot(y_pred)
pt.plot(y_test)
```

polynomial regression

importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

importing the dataset

```python
data=pd.read_csv('Dataset_master.xlsx - RBC in Humans Polynomial Regres.csv')
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
print(x)
```

Training the Linear regression

```python
from sklearn.linear_model import LinearRegression
linear_reg=LinearRegression()
linear_reg.fit(x,y)
```

Training the polynomial linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
linear_reg2=LinearRegression()
linear_reg2.fit(x_poly,y)
```

visualizing the linear results

```python
pt.scatter(x,y,color='red')
pt.plot(x,linear_reg.predict(x))
```

Visualizing the poly reg results

```python
pt.scatter(x,y,color='red')
pt.plot(x,linear_reg2.predict(x_poly))
```

predicting the new result with linear regression

```python
linear_reg.predict([[6.5]])
```

predicitng the new result with poly regressionm

```python
linear_reg2.predict(poly_reg.fit_transform([[5]]))
```

support vector regression

importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

importing the dataset

```python
data_set=pd.read_csv('Dataset_master.xlsx - Salary Data Support Vector regr.csv')
x=data_set.iloc[:,1:-1].values
y=data_set.iloc[:,-1].values
print(x)
```

Feature scaling 

```python
y=y.reshape(len(y),1)
print(y)
```

Training the svr in whole dataset

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
x=sc_x.fit_transform(x)
y=sc_y.fit_transform(y)
print(y)
```

```python
from sklearn.svm import SVR
reg=SVR(kernel='rbf')
reg.fit(x,y)
```

predict the result

```python
sc_y.inverse_transform(reg.predict(sc_x.transform([[6.5]])).reshape(-1,1))
```

Visualising the svr results

```python
pt.scatter(sc_x.inverse_transform(x), sc_y.inverse_transform(y), color = 'red')
pt.plot(sc_x.inverse_transform(x), sc_y.inverse_transform(reg.predict(x).reshape(-1,1)), color = 'blue')
pt.title('(SVR)')
pt.xlabel('Position level')
pt.ylabel('Salary')
pt.show()
```

Decision Tree
A decision tree is a non-parametric supervised learning algorithm, which is utilized for both classification and regression tasks.

Importing the libraies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

Importing the data set

```python
data=pd.read_csv('Dataset_master.xlsx - Salary Data Support Vector regr.csv')
x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
print(y)
```

Training the dataset

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(random_state=21)
reg.fit(x,y)
```

predicting the result

```python
reg.predict([[4]])
```

Visualizing the decision Tree

```python
x_grid=np.arange(min(x),max(x),0.01)
x_grid=x_grid.reshape(len(x_grid),1)

pt.scatter(x,y, color = 'red')
pt.plot(x_grid,reg.predict(x_grid), color = 'blue')
pt.title('Decision Tree Graph')
pt.xlabel('Grade')
pt.ylabel('Salary')
pt.show()
```

Random forest Regression

Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

import the data set

```python
data=pd.read_csv('Dataset_master.xlsx - Salary Data Support Vector regr.csv')
x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
print(y)
```

Training the data set

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=12,random_state=42)
reg.fit(x,y)
```

Predict the result

```python
reg.predict([[5.5]])
```

Visualizing the Random forest

```python
x_grid=np.arange(min(x),max(x),0.01)
x_grid=x_grid.reshape(len(x_grid),1)

pt.scatter(x,y, color = 'red')
pt.plot(x_grid,reg.predict(x_grid), color = 'blue')
pt.title('Random Forest Tree')
pt.xlabel('Grade')
pt.ylabel('Salary')
pt.show()
```
